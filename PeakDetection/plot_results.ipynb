```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
capno_file = "Data/capnostat_slow.txt"
proto_file = "Data/prototype_slow.txt"

c_fp = open(capno_file)
c_len = int(c_fp.readline())-1
capno_data = np.zeros((2, c_len))

for i in range(c_len):
    point = c_fp.readline().replace("\n","").split(',')
    # print(point)
    capno_data[0][i] = float(point[0])
    capno_data[1][i] = float(point[1])

c_fp.close()

p_fp = open(proto_file)
p_len = int(p_fp.readline())-1
proto_data = np.zeros((2, p_len))

for i in range(p_len):
    point = p_fp.readline().replace("\n","").split(',')
    proto_data[0][i] = float(point[0])
    proto_data[1][i] = float(point[1])

p_fp.close()

#Plotting Data after Filtering/Processing
fig, axs = plt.subplots(2)
fig.suptitle("Comparison of Capnostat (red) and Prototype (blue) with 4.3 sensor")
axs[0].plot(capno_data[0], capno_data[1], color="red")
axs[0].set(xlabel="Time (s)")
axs[0].set(ylabel="CO2 (mmHg)")
axs[1].plot(proto_data[0], proto_data[1], color="blue")
axs[1].set(xlabel="Time (s)")
axs[1].set(ylabel="Voltage (V)")

# print(str(len(capno_cleandata)) + " " + str(len(co2_nonco2_diff)))
```
